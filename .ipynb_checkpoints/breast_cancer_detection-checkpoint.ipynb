```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
X.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
def find_high_correlated_features(dataset,threshold):
    cor_cols = set() #Using set for unique columns
    corr_matrix = dataset.corr() #Correlation matrix
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if corr_matrix.iloc[i,j]>threshold:
                colname = dataset.columns[i]
                cor_cols.add(colname)
    return cor_cols
```

```python
correlated_features = find_high_correlated_features(X_train, 0.85)
```

```python
X_train.drop(correlated_features,axis = 1)
X_test.drop(correlated_features,axis=1)
```
